```python
df = pd.read_csv("https://raw.githubusercontent.com/ggarci10/refactored-meme/main/Twitch_game_data%202016-2019.csv")
```

```python
mask = df['Year'] == 2016
 
df_new = pd.DataFrame(df[mask])
 
# print(df_new)

grouped = df_new.groupby("Game")
# sum and average the other information by game
result = grouped.agg({"Hours_streamed": "sum", "Peak_viewers": "max", "Peak_channels": "max", "Avg_viewers": "mean", "Avg_channels": "mean", "Avg_viewer_ratio": "mean"})
print(result)
```
